# Histograms

## All Stellarators

```python
for col in df_0.columns:
    values_0 = df_0[col].values
    values_good_0 = df_good_0[col].values
    values_1 = df_1[col].values
    values_good_1 = df_good_1[col].values
    values_2 = df_2[col].values
    values_good_2 = df_good_2[col].values
    values_3 = df_3[col].values
    values_good_3 = df_good_3[col].values
    values_4 = df_4[col].values
    values_good_4 = df_good_4[col].values
    values_5 = df_5[col].values
    values_good_5 = df_good_5[col].values
    
    if col in ['nfp']:
        vmin = np.min(values_5) - 0.5
        vmax = np.max(values_5) + 0.5
        num = int(round(vmax - vmin + 1))
    
    else:
        vmin = np.min(values_3)
        vmax = np.max(values_3)
        num = 1001
    
    bins = np.linspace(vmin, vmax, num=num)

    plt.hist(values_0, bins, density=True, alpha=0.5, label='dataset', log=True)
    plt.hist(values_1, bins, density=True, alpha=0.5, label='pred_1', log=True)
    plt.hist(values_2, bins, density=True, alpha=0.5, label='pred_2', log=True)
    plt.hist(values_3, bins, density=True, alpha=0.5, label='pred_3', log=True)
    plt.hist(values_4, bins, density=True, alpha=0.5, label='pred_4', log=True)
    plt.hist(values_5, bins, density=True, alpha=0.5, label='pred_5', log=True)

    plt.title(col)
    plt.legend()
    plt.savefig(f'./images/visual_metrics/histograms/{col}.png')
    plt.show()
    plt.close()
```

## Good Stellarators

```python
for col in df_0.columns:
    values_good_0 = df_good_0[col].values
    values_good_1 = df_good_1[col].values
    values_good_2 = df_good_2[col].values
    values_good_3 = df_good_3[col].values
    values_good_4 = df_good_4[col].values
    values_good_5 = df_good_5[col].values
    
    if col in ['nfp']:
        vmin = np.min(values_good_5) - 0.5
        vmax = np.max(values_good_5) + 0.5
        num = int(round(vmax - vmin + 1))
    
    else:
        vmin = np.min(values_good_5)
        vmax = np.max(values_good_5)
        num = 1001
    
    bins = np.linspace(vmin, vmax, num=num)

    plt.hist(values_good_0, bins, density=True, alpha=0.5, label='dataset', log=True)
    plt.hist(values_good_1, bins, density=True, alpha=0.5, label='pred_1', log=True)
    plt.hist(values_good_2, bins, density=True, alpha=0.5, label='pred_2', log=True)
    plt.hist(values_good_3, bins, density=True, alpha=0.5, label='pred_3', log=True)
    plt.hist(values_good_4, bins, density=True, alpha=0.5, label='pred_4', log=True)
    plt.hist(values_good_5, bins, density=True, alpha=0.5, label='pred_5', log=True)

    plt.title(col)
    plt.legend()
    plt.savefig(f'./images/visual_metrics/histograms_good/{col}.png')
    plt.show()
    plt.close()
```
