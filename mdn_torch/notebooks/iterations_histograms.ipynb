# Histograms

## All Stellarators

```python
import matplotlib
# Activate the next line to use LaTeX rendering
matplotlib.rcParams['text.usetex'] = True


for col in df_0.columns:
    values_0 = df_0[col].values
    values_good_0 = df_good_0[col].values
    values_1 = df_1[col].values
    values_good_1 = df_good_1[col].values
    values_2 = df_2[col].values
    values_good_2 = df_good_2[col].values
    values_3 = df_3[col].values
    values_good_3 = df_good_3[col].values
    values_4 = df_4[col].values
    values_good_4 = df_good_4[col].values
    values_5 = df_5[col].values
    values_good_5 = df_good_5[col].values
    
    if col in ['nfp']:
        vmin = np.min(values_5) - 0.5
        vmax = np.max(values_5) + 0.5
        num = int(round(vmax - vmin + 1))
    
    else:
        vmin = np.min(values_3)
        vmax = np.max(values_3)
        num = 1001
    
    bins = np.linspace(vmin, vmax, num=num)

    plt.figure(figsize=(10, 6))

    plt.hist(values_0, bins, density=True, alpha=0.5, label='Before training (uniform sampling)', log=True)
    plt.hist(values_1, bins, density=True, alpha=0.5, label='After the first training session', log=True)
    plt.hist(values_2, bins, density=True, alpha=0.5, label='After the second training session', log=True)
    plt.hist(values_3, bins, density=True, alpha=0.5, label='After the third training session', log=True)
    plt.hist(values_4, bins, density=True, alpha=0.5, label='After the fourth training session', log=True)
    plt.hist(values_5, bins, density=True, alpha=0.5, label='After the fifth training session', log=True)

    # plt.title(col)
    labels = {
        'rc1': r'$\mathrm{R}_{c1}$',
        'rc2': r'$\mathrm{R}_{c2}$',
        'rc3': r'$\mathrm{R}_{c3}$',
        'zs1': r'$\mathrm{Z}_{s1}$',
        'zs2': r'$\mathrm{Z}_{s2}$',
        'zs3': r'$\mathrm{Z}_{s3}$',
        'nfp': r'$\mathrm{n}_{fp}$',
        'etabar': r'$\overline{\eta}$',
        'B2c': r'$B_{2c}$',
        'p2': r'$p_2$',
        'axis_length': r'$\mathrm{Axis\ Length}$',
        'max_elongation': r'$\mathrm{Max\ Elongation}$',
        'min_L_grad_B': r'$\mathrm{Min\ Grad\ B}$',
        'min_R0': r'$\mathrm{Min\ R}_0$',
        'L_grad_grad_B': r'$\mathrm{L Grad\ Grad\ B}$',
        'DMerc_times_r2': r'$\mathrm{D\ Merc\ r}^2$',
        'beta': r'$\beta$',
    }
    plt.xlabel(labels.get(col), fontsize=20)
    plt.legend(fontsize=12)
    plt.ylabel('Density', fontsize=20)
    plt.xticks(fontsize=20)
    plt.yticks(fontsize=20)
    plt.tight_layout()
    plt.savefig(f'../images/visual_metrics/histograms/{col}.pdf')
    plt.show()
    plt.close()
```
